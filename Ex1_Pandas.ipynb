```python
# Passo 1: Download do dataset
!wget -nc http://files.grouplens.org/datasets/movielens/ml-100k.zip

# Passo 2: Extração do dataset
!unzip -n ml-100k.zip

# Passo 3: Carregamento dos dados com Pandas
import pandas as pd




# Carregando os dados de avaliações
ratings = pd.read_csv('ml-100k/u.data', sep='\t', names=['user_id', 'movie_id', 'rating', 'timestamp'])

#Carregando os dados dos usuarios
users = pd.read_csv('ml-100k/u.user', sep='|', names=['user_id', 'age', 'gender', 'occupation', 'zip_code'])

# Carregando os dados dos filmes
movies = pd.read_csv('ml-100k/u.item', sep='|', names=[
    'movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown',
    'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama',
    'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'
], encoding='latin-1')

# Exibindo as primeiras linhas dos DataFrames para verificar o carregamento
print(users.head())
print(ratings.head())
print(movies.head())

```

```python
import pandas as pd

# Carregamento dos dados dos filmes
movies = pd.read_csv('ml-100k/u.item', sep='|', names=[
    'movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'unknown',
    'Action', 'Adventure', 'Animation', 'Children\'s', 'Comedy', 'Crime', 'Documentary', 'Drama',
    'Fantasy', 'Film-Noir', 'Horror', 'Musical', 'Mystery', 'Romance', 'Sci-Fi', 'Thriller', 'War', 'Western'
], encoding='latin-1')

# Remover a coluna 'unknown' que não é relevante
movies = movies.drop(columns=['unknown'])

# Selecionar apenas as colunas de gêneros
genre_columns = movies.columns[5:]

# Criar uma nova coluna 'genre' que irá armazenar os gêneros combinados
movies['genre'] = ''

# Iterar sobre cada linha do DataFrame e combinar os gêneros
for index, row in movies.iterrows():
    genres = []
    for genre in genre_columns:
        if row[genre] == 1:
            genres.append(genre)
    movies.at[index, 'genre'] = ','.join(genres)

# Criar o novo DataFrame condensado
condensed_movies = movies[['movie_id', 'movie_title', 'release_date', 'video_release_date', 'IMDb_URL', 'genre']]

# Exibindo as primeiras linhas do DataFrame resultante
#print(condensed_movies.head())
pd.DataFrame(condensed_movies)

```
